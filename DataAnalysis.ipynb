```python
import pandas as pd
df=pd.read_excel('Candidates_Data/Candidates_data.xlsx')
```

```python
df.head()
```

```python
from datetime import datetime
# create a column with the age of the candidate
df['Age Candidat']=df['Date naissance candidat'].apply(lambda x: datetime.now().year-x.year)
```

```python
df.head()
```

```python
""" Gender information """
values=df['Sexe candidat'].value_counts().array
total_candidates=sum(values)
# percentage of values
pct=[]
for val in values:
    pct.append(val/total_candidates * 100)
print(pct)
```
